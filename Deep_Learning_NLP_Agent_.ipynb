```python
import torch
from transformers import pipeline

# Bind runtime layers straight to the dedicated T4 Cloud GPU
device = 0 if torch.cuda.is_available() else -1
print(f"Using Deep Learning Hardware Target: {'GPU (cuda)' if device == 0 else 'CPU'}")

print("Loading Domain-Specific DistilBERT Ticket Classifier (Fine-Tuned on 61k+ ITSM Logs)...")
itsm_classifier = pipeline(
    "text-classification",
    model="jeremiasdavison/it-support-ticket-classifier",
    device=device
)
print("✅ Production-grade routing engine initialized successfully.")

def run_deep_nlp_inference(raw_text_payload):
    """Processes unstructured ticket logs natively and maps labels safely using case-insensitive keys."""
    print("\nAnalyzing ticket tokens across infrastructure queues...")

    # Process text directly through the fine-tuned sequence classifier layers
    model_output = itsm_classifier(raw_text_payload)[0]

    raw_label = model_output['label']
    confidence_score = model_output['score'] * 100

    # Standardize to uppercase to avoid case-sensitivity failures
    predicted_queue = raw_label.upper().strip()

    # Expanded enterprise action item routing matrix covering all possible dataset strings
    routing_map = {
        "HARDWARE": "🚨 [QUEUE DISPATCH]: Infrastructure Ops / Physical Data Center Engineering Team",
        "SOFTWARE": "💻 [QUEUE DISPATCH]: Application Support / Software DevOps Engineering Team",
        "NETWORK":  "🌐 [QUEUE DISPATCH]: Network Operations Center (NOC) / Security Gateway Team",
        "ACCOUNT":  "🔐 [QUEUE DISPATCH]: Identity & Access Management (IAM) / IT Service Desk",
        "GENERAL":  "📋 [QUEUE DISPATCH]: Tier 1 Service Desk Triage Queue",
        "INCIDENT": "📋 [QUEUE DISPATCH]: Production Infrastructure Triage Group"
    }

    dispatch_action = routing_map.get(predicted_queue, "📋 [QUEUE DISPATCH]: Tier 1 General Triage Line")

    print("\n================== PRODUCTION ITSM ROUTING METRICS ==================")
    print(f"\n[Raw Ticket Payload Ingested]:\n\"{raw_text_payload}\"")
    print(f"\n[Verified Target Queue]: {raw_label}")
    print(f"[True Mathematical Model Confidence]: {confidence_score:.2f}%")
    print(f"\n{dispatch_action}")
    print("======================================================================\n")
```

```python
# =====================================================================
# RUNTIME SINGLE-INPUT DATA INGESTION
# =====================================================================
print("🚀 Custom PyTorch NLP Classifier Pipeline Active.")

# Dynamically capture corporate document data for model inference evaluation
user_payload = input("Paste or type text log to pass through neural network layer: ").strip()

if not user_payload:
    print("❌ Error: Tensor generation aborted. No input tokens detected.")
else:
    # Send text immediately down through the custom deep learning model layers
    run_deep_nlp_inference(user_payload)
```
